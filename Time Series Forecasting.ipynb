```python
## Import standard packages
```

```python
from ast import increment_lineno
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
## Import time series data : Airline passanger traffic
```

```python
data = pd.read_csv("/content/sample_data/airline-passenger-traffic.csv")
data.columns = ['Month','Passengers']
data['Month'] = pd.to_datetime(data['Month'], format = "%Y-%m")
data = data.set_index('Month')
data.head(12)
```

```python
# Time series analysis
```

```python
## Plot time series data
```

```python
data.plot(figsize=(12,4))
plt.legend(loc='best')
plt.title('Airline passenger traffic data')
plt.show(block=False)
```

```python
## Missing value treatment
```

```python
## Mean inputation
```

```python
data = data.assign(Passengers_Mean_Imputation = data.Passengers.fillna(data.Passengers.mean()))
data[['Passengers_Mean_Imputation']].plot(figsize=(12,4))
plt.legend(loc="best")
plt.title('Missing Value: Mean Imputation')
plt.show(block=False)
```

```python
## Linear Interpoaltion
```

```python
data = data.assign(Passengers_Liner_Interpolation = data.Passengers.interpolate(method="linear"))
data[['Passengers_Liner_Interpolation']].plot(figsize=(12,4))
plt.legend(loc="best")
plt.title("Missing value treatment: Linear interpolation")
plt.show(block=False)
```

```python
## use linear interpolation to impute missing value
```

```python
data['Passengers'] = data['Passengers_Liner_Interpolation']
```
